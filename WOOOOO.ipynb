```python
from collections import deque
import bisect
import time
import matplotlib.pyplot as plt
```

```python
def get_kth_element1(k):
    if type(k) != int:
        raise TypeError
    explored_elems = set()
    sorted_unique_elems = [1]
    items_to_explore = deque([1])
    
    while(len(items_to_explore) and len(sorted_unique_elems) < k):
        new = items_to_explore.pop()
        new_items = [new * i for i in [3, 5, 7] ]
        for item in new_items:
            if item not in explored_elems:
                explored_elems.add(item)
                bisect.insort(sorted_unique_elems, item)
                items_to_explore.appendleft(item)
    
    largest_val = sorted_unique_elems[-1]

    while( len(items_to_explore)):
        new = items_to_explore.pop()
        new_items = [new * i for i in [3, 5, 7] ]
        for item in new_items:
            if ((item < largest_val) and (item not in explored_elems)):
                explored_elems.add(item)
                bisect.insort(sorted_unique_elems, item)
                items_to_explore.appendleft(item)
    return sorted_unique_elems[k-1]
```

```python
get_kth_element1(100000)
```

```python
def get_kth_element2(k):
    if type(k) != int:
        raise TypeError
    
    explored_elems = set()
    sorted_unique_elems = [1]
    items_to_explore = deque([1])
    
    while(len(items_to_explore) and len(sorted_unique_elems) < k):
        new = items_to_explore.pop()
        if new not in explored_elems:
            explored_elems.add(new)
            bisect.insort(sorted_unique_elems, new)
            new_items = [new * i for i in [3, 5, 7] ]
            for item in new_items:
                items_to_explore.appendleft(item)
    
    largest_val = sorted_unique_elems[-1]
    
    while( len(items_to_explore)):
        new = items_to_explore.pop()
        if new not in explored_elems:
            explored_elems.add(new)
            bisect.insort(sorted_unique_elems, new)
            new_items = [new * i for i in [3, 5, 7] ]
            for item in new_items:
                if (item < largest_val):
                    items_to_explore.appendleft(item)
    return sorted_unique_elems[k-1]
```

```python
get_kth_element2(100000)
```

```python
def timeit(func, param):
    t2 = time.time()
    func(param)
    t1 = time.time()
    return t2 - t1
```

```python
times = []
for i in range(2500):
    if i % 100 == 0:
        print i
    times.append(timeit(get_kth_element2, i))

plt.plot[range(2500), times]
```
